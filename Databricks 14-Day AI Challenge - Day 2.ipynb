```python
ecommerce_data_2019_nov_df = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv")
ecommerce_data_2019_nov_df.show(5)
```

```python
ecommerce_data_2019_nov_df.createOrReplaceTempView("ecommerce_data_2019_nov_df")
```

```python
%sql
SELECT _c5 as brand, MAX(TRY_CAST(_c6 AS DOUBLE)) AS max_price FROM ecommerce_data_2019_nov_df WHERE TRY_CAST(_c6 AS DOUBLE) > 2573 GROUP BY _c5 ORDER BY max_price DESC
```

```python
result_df = spark.sql("""SELECT _c5, MAX(try_cast(_c6 AS DOUBLE)) AS max_price FROM ecommerce_data_2019_nov_df GROUP BY _c5 ORDER BY max_price DESC""")
result_df.write.csv("/Volumes/workspace/ecommerce/ecommerce_data/ecommerce_data_2019_nov_max_price_by_category.csv", header=True, mode="overwrite")
```
